```python
#Import the needed libraries and packages

import spacy
nlp = spacy.load('en_core_web_sm')

import PyPDF2
```

```python
fp = open('Business_Proposal.pdf', mode = 'rb')
pdf_reader = PyPDF2.PdfFileReader(fp)
pdf_text = []

for p in range(pdf_reader.numPages):
    page = pdf_reader.getPage(p).extractText()
    pdf_text.append(page)
    
pdf_text = "\n".join(pdf_text)
print(pdf_text)
```

```python
stop_words = nlp.Defaults.stop_words
doc = nlp(pdf_text)
new_doc = []

for token in doc:
    if not token.is_stop:
        new_doc.append(token.text)
print(" ".join(new_doc))
```
